```python
# Exericio 3
print("*************** Exercicio 3 ***************")
vef_user = "luiz"
vef_cpf = "5180"
vef_rg = "5556"
again = True

while again != False :
    if again == True:
        usuario = input("Digite o seu usuário:")
        if usuario.lower() == vef_user.lower():
            cpf = input("Digite o seu cpf:")
            if cpf == vef_cpf:
                rg = input("Digite o seu rg:")   
                if rg == vef_rg:
                    print("******** Aceso Liberado ********")
                    print(f"Usuário: {usuario}\nCPF: {cpf}\nRG: {rg}")
                    again = False
                else:
                    print("RG incorreto")     
            else:
                print("CPF incorreto")    
        else: 
            print("Usuário não cadastrado no sistema")
    
    if again != False:
        status = input("Deseja tentar novamente? (S/N)")

        if status.lower() == "n":
            again = False
            print("Sistema encerrado!")
        elif status.lower() == "s":
            again = True
        else:
            again = ''
            print("Valor inválido")

    

```
